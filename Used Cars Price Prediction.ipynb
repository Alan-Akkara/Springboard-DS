```python
def convert_mpg(val):
    if pd.isna(val) or val.strip() == '':
        return np.nan
    try:
        parts = [float(x.strip()) for x in val.split('-')]
        return np.mean(parts)
    except:
        return np.nan 
```

```python
mpg_clean = cars_data['mpg'].apply(convert_mpg)
```

```python
mpg_clean.value_counts().sort_index()
```

```python
cars_data['mpg'] = mpg_clean
```

```python
def save_file(df, filename, folder_path):
    os.makedirs(folder_path, exist_ok=True)
    file_path = os.path.join(folder_path, filename)
    df.to_csv(file_path, index=False)
    print(f"File saved to: {file_path}")
```

```python
datapath = '../data'
save_file(cars_data, 'cars_data_cleaned.csv', datapath)
```
