```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np
```

```python
cars_data = pd.read_csv('cars_data_cleaned.csv')
```

```python
sns.histplot(cars_data['price'], kde=True, bins=30)
plt.title('Price Distribution')
```

```python
sns.histplot(cars_data['mileage'], kde=True, bins=30)
plt.title('Mileage Distribution')
```

```python
sns.histplot(cars_data['mpg'], kde=True, bins=30)
plt.title('MPG Distribution')
```

```python
sns.countplot(y='transmission', data=cars_data, order=cars_data['transmission'].value_counts().index)
```

```python
sns.countplot(y='interior_color', data=cars_data, order=cars_data['interior_color'].value_counts().index)
```

```python
avg_price_year = cars_data.groupby('year')['price'].mean().reset_index()
sns.lineplot(x='year', y='price', data=avg_price_by_year, marker='o')
plt.title('Average Price per Year')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```

```python
sns.boxplot(x='transmission', y='price', data=cars_data)
plt.xticks(rotation=90)
plt.title('Price Distribution by Transmission Type')
plt.show()
```

```python
sns.boxplot(x='drivetrain', y='price', data=cars_data)
plt.xticks(rotation=90)
plt.title('Price Distribution by Drives')
plt.show()
```

```python
sns.boxplot(x='fuel_type', y='price', data=cars_data)
plt.xticks(rotation=90)
plt.title('Price Distribution by Fuel Type')
plt.show()
```

```python
sns.boxplot(x='exterior_color', y='price', data=cars_data)
plt.xticks(rotation=90)
plt.title('Price Distribution by Exteroir Color')
plt.show()
```

```python
sns.boxplot(x='interior_color', y='price', data=cars_data)
plt.xticks(rotation=90)
plt.title('Price Distribution by Interoir Color')
plt.show()
```

```python
sns.boxplot(x='accidents_or_damage', y='price', data=cars_data)
plt.title('Price Distribution by Accidents')
plt.show()
```

```python
sns.scatterplot(x='mpg', y='price', data=cars_data)
```

```python
correlation = cars_data.corr(numeric_only=True)
plt.figure(figsize=(10,6))
sns.heatmap(correlation[['price']].sort_values(by='price', ascending=False), annot=True, cmap='coolwarm')
plt.title('Correlation of Numeric Features with Price')
plt.show()
```

```python
cars_data.groupby('transmission')['price'].mean().sort_values(ascending=False)
```

```python
features = ['price', 'mileage', 'mpg', 'year']
sns.pairplot(cars_data[features + ['fuel_type']], hue='fuel_type', diag_kind='kde')
plt.suptitle("Pairplot of Key Numerical Features", y=1.02)
plt.show()
```
